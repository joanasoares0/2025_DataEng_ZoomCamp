### **Origem do Python**
Python foi projetado para ser uma linguagem que:

### **Diferenças entre Python e outras linguagens**

   #### Exemplo:
   ```c
   // Exemplo em C
   #include <stdio.h>
   int main() {

       return 0;
   }
   ```
   ```python
   # Exemplo em Python
   print("Olá, mundo!")
   ```

2. **Python vs. Java**
   - **Java** exige que você escreva muito código para tarefas simples, devido à sua estrutura baseada em objetos.
   - **Python** permite alcançar o mesmo objetivo com menos código, graças à sua sintaxe enxuta.

   #### Exemplo:
   ```java
   // Exemplo em Java
   public class Main {
       public static void main(String[] args) {
           System.out.println("Olá, mundo!");
       }
   }
   ```
   ```python
   # Exemplo em Python
   print("Olá, mundo!")
   ```

3. **Python vs. JavaScript**
   - **JavaScript** é mais usado no desenvolvimento web, especialmente em navegadores.
   - **Python** é mais versátil e pode ser usado em diversas áreas, como ciência de dados, inteligência artificial e automação.

## **Como Python se caracteriza entre os tipos de linguagem?**

Python possui características que o destacam entre outros tipos de linguagens de programação. Vamos explorar seus principais aspectos e classificações:

### 1. **Linguagem de alto nível**
Python é uma linguagem de **alto nível**, o que significa que sua sintaxe está mais próxima da linguagem humana do que da linguagem de máquina. Isso facilita o aprendizado e reduz o tempo necessário para o desenvolvimento.

#### Exemplo:
```python
# Python: Alto nível e legível
nome = "Maria"
print(f"Bem-vinda, {nome}!")
```

Em linguagens de baixo nível, como Assembly, essa mesma tarefa seria muito mais complexa, exigindo manipulação direta de registradores.

---

### 2. **Interpretada**
Python é uma linguagem **interpretada**, ou seja, seu código é executado linha por linha pelo interpretador, em vez de ser compilado de uma vez. Isso permite:
- Depuração mais rápida.
- Execução imediata do código, ideal para testes.

#### Exemplo:
```python
# Código interpretado é executado diretamente
print("Este código é executado linha por linha!")
```

---

### 3. **Multiparadigma**
Python suporta diferentes paradigmas de programação, incluindo:
- **Programação procedural**: Foco em funções e sequências de instruções.
- **Programação orientada a objetos**: Uso de classes e objetos para organizar o código.
- **Programação funcional**: Uso de funções como "cidadãos de primeira classe" (podem ser atribuídas a variáveis e passadas como argumentos).

#### Exemplo:
```python
# Procedural
def cumprimentar():
    print("Olá, mundo!")
cumprimentar()

# Orientado a objetos
class Pessoa:
    def __init__(self, nome):
        self.nome = nome
    def cumprimentar(self):
        print(f"Olá, {self.nome}!")
pessoa = Pessoa("João")
pessoa.cumprimentar()

# Funcional
cumprimento = lambda nome: f"Olá, {nome}!"
print(cumprimento("Ana"))
```

---

### 4. **Dinamicamente tipada**
Python é uma linguagem **dinamicamente tipada**, o que significa que você não precisa declarar o tipo das variáveis explicitamente. O interpretador identifica o tipo automaticamente.

#### Exemplo:
```python
# Tipagem dinâmica
x = 10       # x é um inteiro
x = "Texto"  # Agora x é uma string
print(x)     # Saída: Texto
```

---

### 5. **Portável**
O código Python é **portável**, o que significa que pode ser executado em diferentes sistemas operacionais (Windows, macOS, Linux) sem alterações significativas, desde que o interpretador Python esteja instalado.

---

Essas características tornam Python uma linguagem extremamente flexível, fácil de usar e adequada para uma ampla gama de aplicações, desde scripts simples até sistemas complexos.

## **Tipos de Dados em Python**

Python possui diversos tipos de dados que permitem armazenar e manipular diferentes tipos de informações. Vamos explorar os principais:

---

### **1. Dados Numéricos**
Python suporta números inteiros, decimais e complexos.

- **Inteiros (`int`)**: Números sem casas decimais.
- **Pontos flutuantes (`float`)**: Números com casas decimais.
- **Números complexos (`complex`)**: Números com parte real e imaginária.

#### Exemplos:
```python
# Inteiros
idade = 25
print("Idade:", idade)

# Pontos flutuantes
altura = 1.75
print("Altura:", altura)

# Números complexos
numero_complexo = 3 + 4j
print("Número Complexo:", numero_complexo)
```

---

### **2. Dados de Texto (Strings)**
Strings são sequências de caracteres, usadas para armazenar texto.

- Definidas entre aspas simples (`'`) ou duplas (`"`).
- Suportam operações como concatenação e repetição.

#### Exemplos:
```python
# String simples
nome = "Alice"
print("Nome:", nome)

# Concatenação
sobrenome = "Silva"
nome_completo = nome + " " + sobrenome
print("Nome completo:", nome_completo)

# Repetição
risada = "Ha" * 3
print("Risada:", risada)
```

---

### **3. Dados Booleanos**
Valores booleanos representam **Verdadeiro** (`True`) ou **Falso** (`False`). São úteis em condições e lógica.

#### Exemplos:
```python
# Booleanos
is_maior_de_idade = True
print("Maior de idade?", is_maior_de_idade)

# Resultado de uma operação
resultado = 10 > 5
print("10 é maior que 5?", resultado)
```

---

### **4. Estruturas de Dados**
Python também possui tipos de dados estruturados que armazenam coleções de valores.

#### Principais:
- **Listas (`list`)**: Coleção ordenada e mutável de itens.
- **Tuplas (`tuple`)**: Coleção ordenada, mas imutável.
- **Conjuntos (`set`)**: Coleção desordenada de itens únicos.
- **Dicionários (`dict`)**: Coleção de pares chave-valor.

#### Exemplos:
```python
# Lista
frutas = ["maçã", "banana", "laranja"]
print("Lista de frutas:", frutas)

# Tupla
coordenadas = (10, 20)
print("Coordenadas:", coordenadas)

# Conjunto
numeros_unicos = {1, 2, 3, 3, 2}
print("Conjunto com números únicos:", numeros_unicos)

# Dicionário
pessoa = {"nome": "Alice", "idade": 25}
print("Dicionário pessoa:", pessoa)
```

---

### **Dica**
Você pode verificar o tipo de uma variável em Python usando a função `type()`.

#### Exemplo:
```python
# Verificar o tipo
x = 10
print("O tipo de x é:", type(x))
```

## **Operações Básicas em Python**

Python permite realizar operações simples e complexas de forma direta. Vamos explorar os principais tipos de operadores.

---

### **1. Operadores Matemáticos**

Os operadores matemáticos em Python incluem:

| Operador | Operação               | Exemplo         | Resultado |
|----------|------------------------|-----------------|-----------|
| `+`      | Adição                 | `5 + 3`         | `8`       |
| `-`      | Subtração              | `5 - 3`         | `2`       |
| `*`      | Multiplicação          | `5 * 3`         | `15`      |
| `/`      | Divisão                | `5 / 2`         | `2.5`     |
| `//`     | Divisão inteira        | `5 // 2`        | `2`       |
| `%`      | Módulo (resto)         | `5 % 2`         | `1`       |
| `**`     | Exponenciação (potência)| `5 ** 2`        | `25`      |

#### Exemplo:
```python
a = 10
b = 3

print("Adição:", a + b)
print("Subtração:", a - b)
print("Multiplicação:", a * b)
print("Divisão:", a / b)
print("Divisão inteira:", a // b)
print("Módulo:", a % b)
print("Exponenciação:", a ** b)
```

---

### **2. Operadores de Comparação**

Os operadores de comparação são usados para comparar valores, retornando `True` ou `False`.

| Operador | Operação                   | Exemplo         | Resultado |
|----------|----------------------------|-----------------|-----------|
| `==`     | Igual a                    | `5 == 5`        | `True`    |
| `!=`     | Diferente de               | `5 != 3`        | `True`    |
| `>`      | Maior que                  | `5 > 3`         | `True`    |
| `<`      | Menor que                  | `5 < 3`         | `False`   |
| `>=`     | Maior ou igual a           | `5 >= 5`        | `True`    |
| `<=`     | Menor ou igual a           | `5 <= 3`        | `False`   |

#### Exemplo:
```python
x = 7
y = 10

print("x é igual a y?", x == y)
print("x é maior que y?", x > y)
print("x é menor ou igual a y?", x <= y)
```

---

### **3. Operadores Lógicos**

Os operadores lógicos são usados para combinar expressões.

| Operador | Operação              | Exemplo               | Resultado |
|----------|-----------------------|-----------------------|-----------|
| `and`    | Retorna `True` se ambas as condições forem verdadeiras | `True and False` | `False`   |
| `or`     | Retorna `True` se pelo menos uma condição for verdadeira | `True or False`  | `True`    |
| `not`    | Inverte o valor lógico | `not True`           | `False`   |

#### Exemplo:
```python
idade = 20
maior_de_idade = idade >= 18
possui_cnh = False

print("Pode dirigir?", maior_de_idade and possui_cnh)
print("Não pode dirigir?", not (maior_de_idade and possui_cnh))
```

---

### **4. Operações com Strings**

Python permite realizar operações úteis em strings:

| Operação       | Exemplo                   | Resultado            |
|-----------------|---------------------------|----------------------|
| Concatenação    | `"Olá" + " Mundo"`        | `"Olá Mundo"`        |
| Repetição       | `"Python " * 3`           | `"Python Python Python"` |
| Índice          | `"Python"[0]`             | `"P"`                |
| Fatiamento      | `"Python"[1:4]`           | `"yth"`              |

#### Exemplo:
```python
texto = "Python"

# Concatenação
print("Olá " + texto)

# Repetição
print(texto * 3)

# Índice e Fatiamento
print("Primeira letra:", texto[0])
print("Parte do texto:", texto[1:4])
```

## **Estruturas Condicionais em Python**

As estruturas condicionais permitem que seu programa tome decisões com base em condições. Em Python, usamos as palavras-chave `if`, `elif` e `else` para criar essas condições.

---

### **1. Estrutura Básica**

A estrutura condicional básica tem a seguinte sintaxe:

```python
if condição:
    # código executado se a condição for verdadeira
elif outra_condição:
    # código executado se a outra condição for verdadeira
else:
    # código executado se nenhuma condição for verdadeira
```

#### Exemplo:
```python
idade = int(input("Digite sua idade: "))

if idade < 18:
    print("Você é menor de idade.")
elif idade >= 18 and idade < 65:
    print("Você é maior de idade.")
else:
    print("Você é idoso.")
```

---

### **2. Condições Simples**

Se você precisa verificar apenas uma condição, pode usar apenas o `if`:

```python
numero = int(input("Digite um número: "))

if numero > 0:
    print("O número é positivo.")
```

---

### **3. Condições Aninhadas**

## **Laços de Repetição em Python**

Os laços de repetição permitem executar um bloco de código várias vezes. Python oferece dois tipos principais de laços:

1. **`for`**: Usado para iterar sobre sequências (listas, strings, ranges, etc.).
2. **`while`**: Executa enquanto uma condição é verdadeira.

---

### **1. Laço `for`**

O laço `for` percorre elementos de uma sequência, como listas ou strings.

#### Sintaxe:
```python
for elemento in sequência:
    # código a ser executado
```

#### Exemplo:
```python
# Iterando sobre uma lista
frutas = ["maçã", "banana", "laranja"]

for fruta in frutas:
    print("Fruta:", fruta)
```

---

### **2. Laço `while`**

O laço `while` continua executando enquanto uma condição for verdadeira.

#### Sintaxe:
```python
while condição:
    # código a ser executado
```

#### Exemplo:
```python
# Contagem regressiva
contador = 5

while contador > 0:
    print("Contador:", contador)
    contador -= 1  # Decrementa o contador
```

---

### **3. Funções Auxiliares**

Python oferece funções que facilitam o uso de loops:

- **`range()`**: Gera uma sequência de números.
  - `range(início, fim, passo)`
- **`enumerate()`**: Retorna o índice e o valor de cada item.
- **`zip()`**: Combina duas ou mais sequências.

#### Exemplo com `range()`:
```python
# Iterando com range
for i in range(1, 6):
    print("Número:", i)
```

#### Exemplo com `enumerate()`:
```python
# Enumerando elementos de uma lista
frutas = ["maçã", "banana", "laranja"]

for i, fruta in enumerate(frutas):
    print(f"Índice {i}: {fruta}")
```

#### Exemplo com `zip()`:
```python
# Combinando listas com zip
nomes = ["Alice", "Bob", "Carol"]
idades = [25, 30, 35]

for nome, idade in zip(nomes, idades):
    print(f"{nome} tem {idade} anos.")
```

---

### **4. Interrompendo e Pulando Itens**

- **`break`**: Encerra o loop imediatamente.
- **`continue`**: Pula para a próxima iteração.

#### Exemplo:
```python
# Usando break
for numero in range(1, 10):
    if numero == 5:
        print("Número 5 encontrado, encerrando...")
        break
    print("Número:", numero)

# Usando continue
for numero in range(1, 10):
    if numero % 2 == 0:
        continue  # Pula números pares
    print("Número ímpar:", numero)
```

---

### **5. Loops Aninhados**

Você pode aninhar loops para trabalhar com dados em várias dimensões.

#### Exemplo:
```python
# Loop aninhado para criar uma matriz
for i in range(1, 4):
    for j in range(1, 4):
        print(f"({i}, {j})", end=" ")
    print()  # Nova linha após cada iteração externa
```

## **Funções em Python**

Funções são blocos de código reutilizáveis que realizam uma tarefa específica. Elas ajudam a organizar o código, evitando repetição e facilitando a manutenção.

---

### **1. O que é uma função?**

Uma função é definida uma vez e pode ser chamada quantas vezes for necessário. Em Python, usamos a palavra-chave `def` para definir uma função.

#### Sintaxe:
```python
def nome_da_funcao(parametros):
    # Bloco de código
    return valor
```

---

### **2. Criando e Chamando Funções**

#### Exemplo:
```python
# Definindo uma função
def cumprimentar():
    print("Olá! Bem-vindo ao Python.")

# Chamando a função
cumprimentar()
```

---

### **3. Funções com Parâmetros**

Parâmetros permitem que você passe informações para a função.

#### Exemplo:
```python
def cumprimentar_pessoa(nome):
    print(f"Olá, {nome}! Bem-vindo ao Python.")

# Chamando a função com um argumento
cumprimentar_pessoa("Maria")
```

---

### **4. Funções com Retorno**

Funções podem devolver um valor usando a palavra-chave `return`.

#### Exemplo:
```python
def somar(a, b):
    return a + b

# Usando o valor retornado
resultado = somar(10, 5)
print("Resultado da soma:", resultado)
```

---

### **5. Parâmetros Padrão**

Você pode definir valores padrão para os parâmetros. Se o argumento não for passado, o valor padrão será usado.

#### Exemplo:
```python
def cumprimentar(nome="visitante"):
    print(f"Olá, {nome}! Bem-vindo ao Python.")

# Chamando a função sem argumento
cumprimentar()

# Chamando a função com argumento
cumprimentar("Carlos")
```

---

### **6. Funções com Vários Argumentos**

1. **`*args`**: Para passar múltiplos argumentos não nomeados.
2. **`**kwargs`**: Para passar múltiplos argumentos nomeados.

#### Exemplo com `*args`:
```python
def somar_todos(*numeros):
    return sum(numeros)

print("Soma total:", somar_todos(1, 2, 3, 4, 5))
```

#### Exemplo com `**kwargs`:
```python
def exibir_dados(**dados):
    for chave, valor in dados.items():
        print(f"{chave}: {valor}")

exibir_dados(nome="Alice", idade=30, cidade="São Paulo")
```

---

### **7. Escopo de Variáveis**

As variáveis dentro de uma função têm **escopo local**, ou seja, só existem enquanto a função está sendo executada.

#### Exemplo:
```python
def funcao():
    x = 10  # Variável local
    print("Dentro da função:", x)

funcao()
# print(x)  # Isso causará erro, pois x não existe fora da função
```

---

### **8. Funções Lambda**

Funções **lambda** são funções anônimas, usadas para criar funções curtas em uma única linha.

#### Exemplo:
```python
# Função lambda para somar dois números
soma = lambda a, b: a + b

print("Soma com lambda:", soma(3, 7))
```

````python
import pandas as pd
# Especificar a planilha
df = pd.read_excel('/content/Financial Sample.xlsx')
print(df.head())
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Exibir as colunas para garantir os nomes corretos
print("Nomes das colunas:")
print(df.columns)

# Análise descritiva básica
print("\nResumo estatístico:")
print(df.describe())

# Histograma para visualizar a distribuição de 'Profit'
plt.figure(figsize=(8, 6))
sns.histplot(df['Profit'], bins=30, kde=True, color='blue')
plt.title('Distribuição de Profit')
plt.xlabel('Profit')
plt.ylabel('Frequência')
plt.show()

# Boxplot para analisar 'Profit' por 'Country'
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Country', y='Profit', palette='Set3')
plt.title('Boxplot de Profit por País')
plt.xlabel('País')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.show()

# Boxplot para analisar 'Profit' por 'Segment'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Segment', y='Profit', palette='Set2')
plt.title('Boxplot de Profit por Segmento')
plt.xlabel('Segmento')
plt.ylabel('Profit')
plt.show()

# Scatterplot usando colunas existentes: Gross Sales e Profit
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Gross Sales', y='Profit', hue='Country', palette='tab10')
plt.title('Relação entre Gross Sales e Profit por País')
plt.xlabel('Gross Sales')
plt.ylabel('Profit')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Pairplot para explorar relações entre variáveis numéricas
numerical_columns = ['Units Sold', 'Gross Sales', 'Profit', 'COGS']
sns.pairplot(df[numerical_columns], diag_kind='kde', corner=True)
plt.show()

# Estatísticas avançadas
print("\nAgrupamento por 'Country' e estatísticas descritivas de Profit:")
country_stats = df.groupby('Country')['Profit'].describe()
print(country_stats)
````

````python
# Construção de novas variáveis
# 1. Lucro por unidade vendida
df['Profit per Unit'] = df['Profit'] / df['Units Sold']

# 2. Margem de lucro (Profit Margin) como porcentagem
df['Profit Margin (%)'] = (df['Profit'] / df['Gross Sales']) * 100

# 3. Nova categorização: Classificar países em alto e baixo desempenho com base no Profit
df['Performance Category'] = df['Profit'].apply(lambda x: 'High' if x > df['Profit'].mean() else 'Low')

print("\nExemplo das novas variáveis:")
print(df[['Profit per Unit', 'Profit Margin (%)', 'Performance Category']].head())

# Análise descritiva das novas variáveis
print("\nResumo estatístico das novas variáveis:")
print(df[['Profit per Unit', 'Profit Margin (%)']].describe())

# Boxplot para visualizar a margem de lucro por país
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Country', y='Profit Margin (%)', palette='coolwarm')
plt.title('Margem de Lucro (%) por País')
plt.xlabel('País')
plt.ylabel('Margem de Lucro (%)')
plt.xticks(rotation=45)
plt.show()

# Detecção de outliers
# Usando IQR (Interquartile Range) para identificar outliers na coluna 'Profit'
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1

# Limites inferior e superior
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("\nLimites para detecção de outliers (Profit):")
print(f"Limite inferior: {lower_bound}")
print(f"Limite superior: {upper_bound}")

# Filtrar outliers
outliers = df[(df['Profit'] < lower_bound) | (df['Profit'] > upper_bound)]
print("\nOutliers identificados:")
print(outliers[['Country', 'Profit']])

# Scatterplot com destaque para os outliers
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Gross Sales', y='Profit', hue='Profit', palette='coolwarm', size='Profit')
plt.axhline(lower_bound, color='red', linestyle='--', label='Limite Inferior')
plt.axhline(upper_bound, color='green', linestyle='--', label='Limite Superior')
plt.title('Scatterplot de Gross Sales vs Profit (com Outliers)')
plt.xlabel('Gross Sales')
plt.ylabel('Profit')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
````

````python
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

cluster_data = df.groupby('Country')[['Profit', 'Gross Sales', 'Units Sold']].mean()

# Escalar os dados para normalizar as variáveis
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_data)

# Criar o gráfico do cotovelo (Elbow Method)
inertia = []
k_range = range(1, min(6, len(cluster_data) + 1))  # Número máximo de clusters não pode exceder o número de países

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plotar o gráfico do cotovelo
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Gráfico do Cotovelo (Elbow Method)')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inércia')
plt.grid()
plt.show()

# Escolher o número ideal de clusters com base no gráfico do cotovelo
optimal_k = 4  # Ajuste esse valor com base no gráfico gerado

# Criar e ajustar o modelo K-Means
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
cluster_data['Cluster'] = kmeans.fit_predict(scaled_data)

# Exibir os resultados
print("\nClusterização por País:")
print(cluster_data)

# Adicionar os clusters de volta ao DataFrame original
df = df.merge(cluster_data['Cluster'], on='Country', how='left')

# Visualizar os clusters com um scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=cluster_data, x='Gross Sales', y='Profit', hue='Cluster', palette='viridis', s=100)
plt.title('Clusters de Países')
plt.xlabel('Média de Gross Sales')
plt.ylabel('Média de Profit')
plt.legend(title='Cluster')
plt.show()

# Exibir estatísticas dos clusters
print("\nEstatísticas dos Clusters:")
print(cluster_data.groupby('Cluster')[['Profit', 'Gross Sales', 'Units Sold']].mean())
````
